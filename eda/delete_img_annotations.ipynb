```python
load_dotenv()
freiburg_dir = os.getenv('FREIBURG_DIR')
```

```python
# script vars
sub_dirs = ['test', 'train', 'valid']
classes_to_remove = ('BEANS', 'OIL', 'VINEGAR', 'SPICES')
```

```python
def process_directories(img_lbl_dirs_path: str):
    """
    Removes files within each directory inside the given parent directory.

    Parameters:
    img_lbl_dirs_path (str): Path to the parent directory containing subdirectories.

    Returns:
    None
    """
    try:
        for data in os.listdir(img_lbl_dirs_path):
            data_dir_path = os.path.join(img_lbl_dirs_path, data)
            # Check if the current path is a directory
            if os.path.isdir(data_dir_path):
                for filename in os.listdir(data_dir_path):
                    if filename.startswith(classes_to_remove):
                        os.remove(filename)
                    else:
                        print(f"Skipped: {filename} is not a directory.")
            else:
                print(f"Skipped: {data_dir_path} is not a directory.")
    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
for dir in sub_dirs:
    img_lbl_dirs_path = os.path.join(freiburg_dir, dir)
    process_directories(img_lbl_dirs_path)      
        
```
